This analysis for Apple so we only need Apple's data

```python
import pandas as pd


path = r'C:\\Users\\omerf\\OneDrive\\Masaüstü\\Data Analysis Project\\S&P Stock Market\\individual_stocks_5yr\\AAPL_data.csv'

apple = pd.read_csv(path)
```

```python
apple["date"] = pd.to_datetime(apple["date"])
apple.head()
```

```python
#Find percentage changes using the pct_change function (multiply by 100)
apple["Daily Return (in %)"] = apple["close"].pct_change() * 100
```

```python
apple.fillna(0).head()
```

```python
#This plot is static it can't be seen in the github repo
import plotly.express as px
px.line(apple, x = "date", y = "Daily Return (in %)")
```

```python
#This is for github it is not as good as the first one
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.xlabel("Date")
plt.ylabel("Daily Return (in %)")
plt.title("Closing Price Change in Apple")
plt.plot("date", "Daily Return (in %)", data = apple)
```

Negative numbers mean it is lower than last day, positive numbers mean it is higher than last day. Most increase at 24 April 2014, almost %8 percent. Most decrease at Jan 28, 2014, almost %7 percent.
