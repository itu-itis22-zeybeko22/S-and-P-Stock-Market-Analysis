This analysis for Apple Data too

```python
import pandas as pd


path = r'C:\\Users\\omerf\\OneDrive\\Masaüstü\\Data Analysis Project\\S&P Stock Market\\individual_stocks_5yr\\AAPL_data.csv'

apple = pd.read_csv(path)
```

```python
apple["date"] = pd.to_datetime(apple["date"])
apple.head()
```

```python
apple.set_index("date", inplace=True)
apple.head()
```

```python
apple["close"].resample("ME").mean()
```

```python
#Monthly resampling
apple["close"].resample("ME").mean().plot()
```

```python
#Yearly resampling
apple["close"].resample("YE").mean().plot()
```

```python
#Quarter resampling
apple["close"].resample("QE").mean().plot()
```

After 2016 in every quarter Apple Stock Prices have increasing trends same for yearly changes. But in monthly resampling you can see changes more clearly. It doesn't have stable increase in trends.
